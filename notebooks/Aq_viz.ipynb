```python
# Assuming df is your original DataFrame
numeric_columns = df.select_dtypes(include=['float64', 'int64'])

# Create a new DataFrame with only numeric columns
numeric_df = df[numeric_columns.columns]

# Display the new DataFrame
print(numeric_df.head())
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
df = df.drop('PM2.5', axis=1)
```

```python
#df=df.drop('Month',axis=1)
df=df.drop('Date',axis=1)
```

```python
# Sum up occurrences of each city across all one-hot encoded columns
city_distribution = df.iloc[:, 14:40].sum()

# Plot the city distribution
plt.figure(figsize=(12, 6))
city_distribution.plot(kind='bar', color='skyblue')
plt.title('City Distribution')
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# List of chemical parameters
chemical_parameters = ['PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']

# Plot scatter plots for each chemical parameter against AQI
plt.figure(figsize=(15, 15))
for parameter in chemical_parameters:
    plt.subplot(4, 3, chemical_parameters.index(parameter) + 1)
    sns.scatterplot(x=parameter, y='AQI', data=df, alpha=0.5)
    plt.title(f'AQI vs {parameter}')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[chemical_parameters + ['AQI']])
plt.title('Box Plot for Chemical Parameters and AQI')
plt.xticks(rotation=45)
plt.show()

```

```python
df1=pd.read_csv('AQI_Pre.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Bar plot for average AQI across different months
plt.figure(figsize=(12, 6))
sns.barplot(x='Month', y='AQI', data=df1)
plt.title('Average AQI Across Months')
plt.xlabel('Month')
plt.ylabel('Average AQI')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert 'Date' column to datetime format
df1['Date'] = pd.to_datetime(df1['Date'])

# Extracting the year from the 'Date' column
df1['Year'] = df1['Date'].dt.year

# Line plot for AQI variation across months over the years
plt.figure(figsize=(14, 8))
sns.lineplot(x='Month', y='AQI', hue='Year', data=df1, ci=None)
plt.title('AQI Variation Across Months Over the Years')
plt.xlabel('Month')
plt.ylabel('AQI')
plt.legend(title='Year', loc='upper right', bbox_to_anchor=(1.15, 1))
plt.show()
```
